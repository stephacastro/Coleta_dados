# Método Apply e Ordenação de DataFrames - Youtube
* Método apply
* Método sort_values

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('C:\\Series_Temporais\\UDEMY_TSA_FINAL\\titanic_train.csv')
```

```python
df.head()
```

```python
# Aplica sobre o dataframe uma função - aplicado em uma coluna especifica - passa como parametro o nome da função e ele pega
# cada elemento da coluna e aplicar a função em cada elemento

df['Age'].apply(np.sqrt)
```

```python
def changeSurvived(elemento):
    if elemento == 0:
        return 'No'
    else:
        return 'Yes'

df['Survived'] = df['Survived'].apply(changeSurvived)
```

```python
df.head()
```

```python
df['Fare'].dtype
```

```python
def changeFare(elemento):
    return '${:.2f}'.format(elemento)

df['Fare'] = df['Fare'].apply(changeFare)
```

```python
df['Fare']
```

```python
df.head()
```

# Ordenação

```python
# Ordena por pelo parametro que for passado e por padrão é em ordem crescente
df_sort = df.sort_values(by='Age')
```

```python
df_sort
```

```python
df_sort.dropna()
```

```python
df_sort = df.sort_values(by='Age', ascending=False)
```
