# Subplots

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
df = pd.read_csv('C:\\Series_Temporais\\UDEMY_TSA_FINAL\\life_expectancy.csv')
```

```python
df.head()
```

```python
# Conferindo os valores unicos do meu dataset
df['Country'].unique()
```

```python
# Excluindo a coluna Unnamed: 0
df.drop('Unnamed: 0', axis=1, inplace=True)
```

```python
df.head(20)
```

```python
# Filtrando os dados
df_brasil = df[df['Country'] == 'Brazil']
```

```python
df_brasil
```

```python
# Filtrando os dados
df_eua = df[df['Country'] == 'EUA']
```

```python
df_eua
```

### Subplots

```python
# Retorna uma figura e um axes
figure, axes = plt.subplots(nrows=1, ncols=2, figsize=(14,5))


axes[0].plot(df_brasil['Year'], df_brasil['Life Expectancy'], marker='o')
axes[0].set_xlabel('Ano')
axes[0].set_ylabel('Expectativa de Vida')
axes[0].set_title('Dados do Brasil')

axes[1].plot(df_eua['Year'], df_eua['Life Expectancy'], marker='o')
axes[1].set_xlabel('Ano')
axes[1].set_ylabel('Expectativa de Vida')
axes[1].set_title('Dados dos Estados Unidos')
```

```python
figure, axes = plt.subplots(nrows=2, ncols=1, figsize=(12,8))


axes[0].plot(df_brasil['Year'], df_brasil['Life Expectancy'], marker='o')
axes[0].set_xlabel('Ano')
axes[0].set_ylabel('Expectativa de Vida')
axes[0].set_title('Dados do Brasil')

axes[1].plot(df_eua['Year'], df_eua['Life Expectancy'], marker='o')
axes[1].set_xlabel('Ano')
axes[1].set_ylabel('Expectativa de Vida')
axes[1].set_title('Dados dos Estados Unidos')

plt.tight_layout()
```
