# Plotagem Básica II & Subplots
* Plotando mais de um gráfico em um mesmo axes
* Definindo legendas
* Definindo os valores contidos nos eixos
* Subplots

```python
import numpy as np
import matplotlib.pyplot as plt

# Para que os graficos fiquem dentro de uma propria linha
%matplotlib inline
```

```python
x = np.arange(0,10,0.01)

y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.sin(x) + np.cos(3*x)
```

```python
# Plotando mais de um gráfico em um mesmo axes
plt.plot(x, y1, label='Sin(x)')
plt.plot(x, y2, label='Cos(x)')
plt.plot(x, y3, label='Sin(x) + Cos(3x)')

# Adicionando legenda
plt.legend(loc='upper left')

# essa função permiti definir todos os valores que se refere ao eixo horizontal
plt.xticks(range(11))
plt.yticks([-2,0,2])

plt.grid(True)
```

### Subplots

```python
plt.figure(figsize=(12,4))

plt.subplot(1,3,1)
plt.plot(x,y1)
plt.subplot(1,3,2)
plt.plot(x,y2)
plt.subplot(1,3,3)
plt.plot(x,y3)
plt.tight_layout()
```

```python
y4 = np.cos(x) + np.sin(3*x)
```

```python
plt.subplot(2,2,1)
plt.plot(x,y1)
plt.subplot(2,2,2)
plt.plot(x,y2)
plt.subplot(2,2,3)
plt.plot(x,y3)
plt.subplot(2,2,4)
plt.plot(x,y4)

plt.tight_layout()
```
