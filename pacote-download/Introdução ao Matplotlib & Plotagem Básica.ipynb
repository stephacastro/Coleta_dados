# Introdução ao Matplotlib & Plotagem Básica

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
array = np.arange(1,5)

plt.plot(array)
```

```python
x = np.arange(0,10,0.2)
y = np.sin(x)
```

```python
plt.plot(x,y)

plt.title('Seno')
plt.xlabel('X')
plt.ylabel('Y')

plt.grid(True)
```
