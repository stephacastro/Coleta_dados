# Customizing Plots Created With Pandas - Estudos Part2

```python
import pandas as pd
import numpy as np
%matplotlib inline
```

```python
df2 = pd.read_csv('df2.csv')
```

```python
df2
```

```python
# personalizando o grafico de linha
# ls='--' estilo da linha
# c='red'cor da linha
# lw=3 espessura da linha
df2['c'].plot.line(figsize=(15,3), ls='-.', c='red', lw=3)
```

```python
# adicionando uma legenda ao grafico e rotulos ao eixo x e y
title = 'MY PLOT TITLE'
xlabel = 'MY X DATA'
ylabel = 'MY Y DATA'
```

```python
ax = df2['c'].plot.line(figsize=(15,3), ls='-.', c='red', lw=3)
ax.set(xlabel=xlabel, ylabel=ylabel)
ax.set(title=title)
```

```python
# personalizando a posio da legenda 
# loc=0 o pandas tenta encontrar a melhor posio para legenda
# bbox_to_anchor=(1.0,4.0) posiciona a legenda fora do grafico
ax = df2.plot.line(figsize=(15,5))
ax.legend(loc=0, bbox_to_anchor=(1.0,0.5))
```
