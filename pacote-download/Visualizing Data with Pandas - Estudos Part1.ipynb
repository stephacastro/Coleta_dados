# Visualizing Data with Pandas - Estudos Part1

```python
import pandas as pd
%matplotlib inline
```

```python
df = pd.read_csv('df1.csv', index_col=0)
```

```python
df.head()
```

```python
df2 = pd.read_csv('df2.csv')
```

```python
df2.head()
```

```python
# Plotando um histograma
df['A'].plot.hist()
```

```python
# personalizando o histograma
# bins=100 - parametro que personaliza a quantidade de barras no hist
# edgecolor='k' - parametro que contorna as barras do hist
# autoscale(enable=True, axis='both', tight=True) - parametro que alinha a barra com o tamanho do hist
df['A'].plot.hist(bins=10, edgecolor='k',figsize=(10,5)).autoscale(enable=True, axis='both', tight=True)
```

```python
# grid=True - parametro que coloca grade no hist
df['A'].plot.hist(grid=True, figsize=(10,5))
```

```python
df2.head()
```

```python
# grafico de barras 
# stacked=True - parametro que empilha os dados do grafico
df2.plot.bar(stacked=True, figsize=(10,5))
```

```python
# grafico de barras na horizontal
# stacked=True - parametro que empilha os dados do grafico
df2.plot.barh(stacked=True, figsize=(10,5))
```

```python
# Grafico de linha
# y='a' - parametro que define a coluna que quer pegar os dados
# figsize=(10,5) -  parametro que define o tamanho do grafico caso no queira o padro
# lw=4 - parametro que engrossa a linha do grafico
df2.plot.line(y='a', figsize=(10,5), lw=4)
```

```python
# Grafico de linha
# y=['a','b','c'] - parametro que define a coluna que quer pegar os dados
# figsize=(10,5) -  parametro que define o tamanho do grafico caso no queira o padro
# lw=4 - parametro que engrossa a linha do grafico
df2.plot.line(y=['a','b','c'], figsize=(10,5), lw=4)
```

```python
# grafico de area padro
df2.plot.area(figsize=(15,5))
```

```python
# grafico de area personalizado
# alpha=0.4 - parametro que trs transparencia para o grafico
df2.plot.area(stacked=False, alpha=0.4, figsize=(15,5))
```

```python
# grafico de dispersso
# cmap='coolwarm' - parametro que deixa o grafico com cores quentes
df.plot.scatter(x='A', y='B',c='C', cmap='coolwarm', figsize=(15,5))
```

```python
# grafico de caixa
df2.plot.box(figsize=(15,5))
```

```python
# grafico de KDE
df2.plot.kde(figsize=(15,5))
```
